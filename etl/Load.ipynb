```python
import os
import json
from pprint import pprint
```

# 2021

```python
with open(os.path.join('data', '2021', 'trip_count_summary.json')) as jsonfile:
    trip_count_summary_2021 = json.load(jsonfile)

with open(os.path.join('data', '2021', 'avg_dur_summary.json')) as jsonfile:
    avg_dur_summary_2021 = json.load(jsonfile)

with open(os.path.join('data', '2021', 'stations_summary.json')) as jsonfile:
    stations_summary_2021 = json.load(jsonfile)
```

# 2022

```python
with open(os.path.join('data', '2022', 'trip_count_summary.json')) as jsonfile:
    trip_count_summary_2022 = json.load(jsonfile)

with open(os.path.join('data', '2022', 'avg_dur_summary.json')) as jsonfile:
    avg_dur_summary_2022 = json.load(jsonfile)

with open(os.path.join('data', '2022', 'stations_summary.json')) as jsonfile:
    stations_summary_2022 = json.load(jsonfile)
```

```python
# # Check all 2021 stations appear in 2022 stations.
# all(elem in stations_summary_2022.keys() for elem in stations_summary_2021.keys())
```

# Merge

```python
data_summary = [{'trip_count': {'2021':trip_count_summary_2021, '2022':trip_count_summary_2022},
                'trip_avg_dur': {'2021':avg_dur_summary_2021, '2022':avg_dur_summary_2022},
                'stations': {'2021':stations_summary_2021, '2022': stations_summary_2022}}]

with open(os.path.join('data', 'data_summary.json'), 'w') as outfile:
    json.dump(data_summary, outfile)
```
