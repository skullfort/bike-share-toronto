```python
# Dependencies
import requests
from pprint import pprint

import os
import pandas as pd
import numpy as np
```

# Station API

```python
base_url = "https://ckan0.cf.opendata.inter.prod-toronto.ca"
url = base_url + "/api/3/action/package_show"
```

```python
params = {"id": "bike-share-toronto"}
package = requests.get(url, params=params).json()
pprint(package)
```

```python
bike_share_json = requests.get(package["result"]["resources"][0]['url']).json()
pprint(bike_share_json)
```

```python
# Access the json data with the name 'station_information'.
station_information_json = requests.get(bike_share_json['data']['en']['feeds'][2]['url']).json()
pprint(station_information_json)
```

## 2021

```python
# Download the monthly CSV files and place them in the same folder (data/2021/) as the output CSV file.
# Convert them into one DataFrame.
df_li = []
for i in np.arange(1,13):
    # ISO-8859-1 encoding must be used to avoid error in reading the CSV files.
    df = pd.read_csv(os.path.join('data', '2021', f'Bike share ridership 2021-{i:02d}.csv'), encoding='ISO-8859-1')
    df_li.append(df)
df = pd.concat(df_li, axis=0, ignore_index=True)

# `Trip Id` is split into two separate columns during import due to the way the headers are decoded, but both can be dropped.
df.drop(columns=[df.columns[0], df.columns[-1]], inplace=True)

# Export the DataFame to a CSV file.
df.to_csv(os.path.join('data', '2021', 'bst_ridership.csv'), index=False)
```

## 2022

```python
# Download the monthly CSV files and place them in the same folder (data/2022/) as the output CSV file.
# Convert them into one DataFrame.
df_li = []
for i in np.arange(1,13):
    df = pd.read_csv(os.path.join('data', '2022', f'Bike share ridership 2022-{i:02d}.csv'))
    df_li.append(df)
df = pd.concat(df_li, axis=0, ignore_index=True)

# Each trip is marked by a unqiue `Trip Id`, which has no special meaning, so it can be dropped.
df.drop(columns=['Trip Id'], inplace=True)

# Export the DataFame to a CSV file.
df.to_csv(os.path.join('data', '2022', 'bst_ridership.csv'), index=False)
```
